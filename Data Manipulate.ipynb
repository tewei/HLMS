```python
#j = 0
#for i in raw_list[0]:
#    print j, i,
#    j+=1

def stable(relation, half_life):
    if ((relation == "=" or relation == ">" or relation == ">=") and half_life > 0.5) or ((relation == ">" or relation == ">=") and half_life == 0.5):
        return 1
    else:
        return 0

final_list = [[i[0], i[28], i[10], i[12], i[13], i[14], i[15], i[6], i[7], i[8]] for i in raw_list]

for i in range(len(final_list)):
    if i == 0:
        final_list[i].append("STABLE")
    else:
        final_list[i][5] = float(final_list[i][5])
        final_list[i].append(stable(final_list[i][4], final_list[i][5]))
        if (final_list[i][4] == "<" and final_list[i][5] > 0.5) or (final_list[i][4] == ">" and final_list[i][5] < 0.5):
            print final_list[i][4], final_list[i][5], "Cannot Categorize", i
        
    #print final_list[i]

end = 0
while(not end):
    end = 1
    for i in range(len(final_list)):
        for j in range(i+1, len(final_list)):
            if(final_list[i][2] == final_list[j][2]):
                print "Same SMILE %d %d" % (i, j)
                print "%s" % final_list[i][2]
                print "%s" % final_list[j][2]
                del final_list[i]
                end = 0
                break

#print final_list
```
